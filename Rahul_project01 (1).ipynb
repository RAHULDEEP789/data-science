# Boxplots

```python
numerical_columns = data.select_dtypes(include=['int64', 'float64']).columns

# Plot box plots for all numerical columns
plt.figure(figsize=(12, 8))
data[numerical_columns].boxplot()
plt.title('Boxplot for All Columns')
plt.xticks(rotation=45)  # Rotate x-axis labels for better visibility
plt.show()
```

# Count Plot

```python
column_name = 'ID'

# Create count plot
plt.figure(figsize=(8, 6))
sns.countplot(x=column_name, data=data)
plt.title('Count Plot for {}'.format(column_name))
plt.xlabel(column_name)
plt.ylabel('Count')
plt.show()
```

```python
for col in data.columns:
    Q1 = data[col].quantile(0.25)
    Q3 = data[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_limit = Q1 - 1.5 * IQR
    upper_limit = Q3 + 1.5 * IQR
    outliers = data[(data[col] < lower_limit) | (data[col] > upper_limit)].index
    print(f"Outliers in column '{col}': {outliers.tolist()}")
```

#  frequency distribution

```python
column_name = 'ID'

# Create frequency distribution plot (histogram)
plt.figure(figsize=(8, 6))
plt.hist(data[column_name], bins=10, edgecolor='black')
plt.title('Frequency Distribution Plot for {}'.format(column_name))
plt.xlabel(column_name)
plt.ylabel('Frequency')
plt.show()
```

# Frequency Distribution Observation

The diagram shows a frequency distribution plot for a variable called "ID".

The x-axis of the plot shows the possible values of the ID variable, ranging from 0 to 1.2.

The y-axis of the plot shows the frequency of each value, with the highest frequency being around 0.4.

The plot appears to be a bimodal distribution, meaning that there are two peaks in the distribution. This suggests that there may be two distinct groups of IDs in the data.

The left peak of the distribution is wider than the right peak, suggesting that there are more IDs in the lower range of values.

# Scatter plot

```python
x_column = 'ID'
y_column = 'Salary'

# Create scatter plot
plt.figure(figsize=(8, 6))
plt.scatter(data[x_column], data[y_column])
plt.title('Scatter Plot between {} and {}'.format(x_column, y_column))
plt.xlabel(x_column)
plt.ylabel(y_column)
plt.grid(True)
plt.show()
```

# Scatter Plot Observation

There is a positive correlation between ID and salary. This means that as the ID number increases, the salary also tends to increase.
The data points are not all tightly clustered together, which suggests that there is some variability in salaries for each ID number.
There are a few outliers, which are data points that fall far away from the main trend of the data.
The salary range is quite large, with some people earning much more than others.

# Bar Plot

```python
numerical_column = 'Salary'
categorical_column = 'Gender'

# Create bar plot
plt.figure(figsize=(8, 6))
sns.barplot(x=categorical_column, y=numerical_column, data=data, estimator=sum)
plt.title('Mean {} by {}'.format(numerical_column, categorical_column))
plt.xlabel(categorical_column)
plt.ylabel('Mean {}'.format(numerical_column))
plt.show()
```

# Sacked Bar Plot


```python
stacked_data = data.groupby(['Degree', 'Gender']).size().unstack(fill_value=0)

# Create stacked bar plot
stacked_data.plot(kind='bar', stacked=True, figsize=(10, 6))
plt.title('Stacked Bar Plot of Gender by Designation')
plt.xlabel('Designation')
plt.ylabel('Count')
plt.xticks(rotation=90)
plt.legend(title='Gender')
plt.show()
```

# Stacked Bar Plot Observation
1.There are eight designations/qualifications in total: B.Tech/B.E., M.Sc. (Tech.)., M.Tech./M.E., MCA, and four unspecified designations (1-4).

2.The x-axis represents the count of individuals in each category, while the y-axis represents the different designations/qualifications.

3.The plot is divided into two stacked bars for each category, representing the gender distribution: one for females (light color) and one for males (dark color).

4.The highest count of individuals is observed in the B.Tech/B.E. category, with approximately 3500 individuals, followed by M.Tech./M.E. with around 3000 individuals.

5.The gender distribution appears to be relatively balanced across all categories, with a slight predominance of males in some categories, such as B.Tech/B.E. and M.Tech./M.E.

6.The unspecified designations (1-4) have lower counts compared to B.Tech/B.E. and M.Tech./M.E., and the gender distribution appears to be relatively balanced in these categories as well.
